```python
#def get_image(image_name):
#    path = ('D://Wallpapers/{}')
#    return path.format(image_name)
import cv2
import matplotlib.pyplot as plt

def draw_grid(image, grid_size):
    height = image.shape[0]
    width = image.shape[1]
    
    x = grid_size
    while x < width :
        cv2.line(image, (x, 0), (x, height), (0,255,0), 2)
        x += grid_size
    y = grid_size
    while y < height:
        cv2.line(image, (0, y), (width, y), (0,255,0), 2)
        y += grid_size

        
def main(path, grid_size):
    image = cv2.imread(path)
    # image_copy = image
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    draw_grid(image, grid_size)
    
    plt.imshow(image)
    plt.show()
    
```

```python
#def get_image(image_name):
#    path = ('D://Wallpapers/{}')
#    return path.format(image_name)
import cv2
import matplotlib.pyplot as plt
import numpy as np

def draw_grid(image, grid_shape):
    
    h, w, _ = image.shape
    rows, cols = grid_shape
    dy, dx = h/rows, w/cols
    
    # Draw vertical lines
    for x in np.linspace(start=dx, stop=w-dx, num=cols-1):
        x = int(round(x))
        cv2.line(image, (x, 0), (x, h), color=(0,255,0), thickness=2)
    # Draw horizontal lines
    for y in np.linspace(start=dy, stop=h-dy, num=rows-1):
        y = int(round(y))
        cv2.line(image, (0, y), (w, y), color=(0,255,0), thickness=2)
        

def main(path, grid_shape):
    image = cv2.imread(path)
    # image_copy = image
    image = cv2.cvtColor(image, cv2.COLOR_BGR2BGR)

    draw_grid(image, grid_shape)
    
    cv2.imwrite("D://Wallpapers/Cara_de_sachiel_lastimada_grid.png", image)
```
